```python

# ── SEED VARIETY DATA ────────────────────────────────────────
# Dominant maize variety planted in Uasin Gishu each year
# Based on KEPHIS records + CIMMYT adoption studies

seed_data = {
    "Year": [2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,
             2015,2016,2017,2018,2019,2020,2021,2022,2023,2024],

    "Dominant_Variety": [
        "H614D","H614D","H614D","H614D","H614D",        # 2005–2009
        "H614D","H614D",                                  # 2010–2011
        "H614D","DH04","DH04",                            # 2012–2014
        "DH04","DH04","DK8031","DK8031","DK8031",        # 2015–2019
        "DK8031","DK8031","DK8031","H6213","H6213"       # 2020–2024
    ],

    "Yield_Potential_Min_tHa": [
        6,6,6,6,6,   # H614D
        6,6,
        6,7,7,        # H614D, DH04, DH04
        7,7,8,8,8,   # DH04, DH04, DK8031 x3
        8,8,8,7,7    # DK8031 x3, H6213 x2
    ],

    "Yield_Potential_Max_tHa": [
        8,8,8,8,8,   # H614D
        8,8,
        8,9,9,        # H614D, DH04, DH04
        9,9,10,10,10, # DH04, DH04, DK8031 x3
        10,10,10,9,9  # DK8031 x3, H6213 x2
    ],

    "Maturity_Months": [
        6.5,6.5,6.5,6.5,6.5,
        6.5,6.5,
        6.5,7.0,7.0,
        7.0,7.0,6.5,6.5,6.5,
        6.5,6.5,6.5,7.0,7.0
    ],

    "Source": ["KEPHIS/CIMMYT"] * 20
}

df_seeds = pd.DataFrame(seed_data)

# Calculate yield potential midpoint — useful ML feature
df_seeds["Yield_Potential_Mid"] = (
    df_seeds["Yield_Potential_Min_tHa"] + 
    df_seeds["Yield_Potential_Max_tHa"]
) / 2

print("✓ Seed variety data created")
print(df_seeds.to_string(index=False))

# Save
df_seeds.to_csv("../data/processed/seeds_clean.csv", index=False)
print("\n✓ Saved seeds_clean.csv")
```
